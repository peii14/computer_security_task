```python
def encrypt(permutation, word_replacements, message):
    # Step 1: Permutation
    encrypted_text = ''.join([permutation.get(ch, ch) for ch in message])

    # Step 2: Replacing Words
    for word, substitute in word_replacements.items():
        # Replace the word when it's not at the junction of two words
        encrypted_text = encrypted_text.replace(" " + word + " ", " " + substitute + " ")
        # Handle the case where the word is at the beginning or end of the text
        if encrypted_text.startswith(word + " "):
            encrypted_text = encrypted_text.replace(word + " ", substitute + " ", 1)
        if encrypted_text.endswith(" " + word):
            encrypted_text = encrypted_text.rreplace(" " + word, " " + substitute, 1)
        # Handle the case where the word is part of another word
        encrypted_text = encrypted_text.replace(word, substitute)

    return encrypted_text
```

```python
permutation = {
    'a': 'z',
    'b': 'y',
    'c': 'x',
}
word_replacements = {
    'the': '2',
    'and': '3'
}
message = "rather and the cat"
encrypt(permutation, word_replacements, message)
```
